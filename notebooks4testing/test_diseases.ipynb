```python
import pandas as pd

diseases = pd.read_csv("../diseases.csv")
dhis_diseases = diseases.loc[:,"DHIS2"]
```

```python
dhis_diseases = dhis_diseases.to_list()
```

```python

def create_diseases_dict():
    gnu_diseases = list()
    diseases = dict()
    for i in dhis_diseases:
        pathologies = Model.get(model)
        records = pathologies.find([('name', 'ilike', f'%{i}%')])
        diseases[i] = records
    return diseases
```

```python
diseases = create_diseases_dict()
```

```python
diseases
```

```python
def get_as_df(disiseases_dict):
    df = pd.DataFrame(columns=["dhis2_name","gnu_health_name", "gnu_health_code"])
    n = 0
    for k,v in disiseases_dict.items():
            for i in v:
                df.loc[n] = [k,i.name,i.code]
                n = n + 1
    return df
```

```python
df = get_as_df(diseases)
```
